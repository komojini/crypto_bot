```python

# cryptobot/main.py


from crypto_bot.binance.binance_client import BinanceClient
from crypto_bot.binance.message_handler import BinanceWSMessageHandler
from crypto_bot.utils.util import repeat_running_until_keyboard_interrupt
from crypto_bot.binance import message_handler as mh 
from crypto_bot import market_data as m

import os
from decimal import Decimal
from datetime import datetime
from queue import Queue
from multiprocessing import Process


streams = [
    "trbusdt@aggTrade",
]

message_handler = BinanceWSMessageHandler(
    # callback=lambda msg: print(msg),
)
binance_client = BinanceClient(
    message_handler=message_handler
)


spot_agg_trade_queue = Queue(maxsize=1000)
um_agg_trade_queue = Queue(maxsize=1000)


def fn():
    os.system("clear")
    trb = m["binance"]["TRBUSDT"]
    trb_um = trb["um"]
    trb_spot = trb["spot"]

    trb_spot_agg_trade = trb_spot["aggTrade"]
    trb_um_agg_trade = trb_um["aggTrade"]


    spot_agg_trade_queue.put(trb_spot_agg_trade)
    um_agg_trade_queue.put(trb_um_agg_trade)

    avg_spot_agg_trade_volume = sum(
        [Decimal(x["q"]) for x in spot_agg_trade_queue.queue]
    )
    avg_um_agg_trade_volume = sum(
        [Decimal(x["q"]) for x in um_agg_trade_queue.queue]
    )

    os.system("clear")

    print(f"""
TRB/USDT
Spot Agg Trade Volume: {avg_spot_agg_trade_volume}
UM Agg Trade Volume  : {avg_um_agg_trade_volume}

Spot Price           : {trb_spot_agg_trade["p"]}
UM Price             : {trb_um_agg_trade["p"]}
""")






def run(): 
    repeat_running_until_keyboard_interrupt(
        fn=fn,
        interval=1,
    )

if __name__ == "__main__":


    binance_client.start_stream(
        spot_streams=streams,
        um_streams=streams,
    )

    run()
```

```python
process.join()
process.terminate()
```
